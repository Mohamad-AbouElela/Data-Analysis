```python
# import required liberaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import sys
from datetime import time
%matplotlib inline 
```

```python
# read data
df = pd.read_csv('Ironman World Championship.csv', encoding= 'unicode_escape')
print (df.shape)
df.head()
```

```python
#  explore variables types
df.info()
```

```python
# explore variables values
df.describe(include = 'all')
```

```python
# check for null values
df.isnull().sum()
```

### Dataset Cleaning Requirments:
1. Data set is almost clean, no null values
2. Change the time to be in minutes for easier processing
3. Divide the data into female and male dataframes to compare each gender separtely
4. Create seprate dataframes for First, Second, and Thirds positions


```python
# Change Time to Datetime object
df['Time'] = pd.to_datetime(df['Time'], format ='%H:%M:%S').dt.time
```

```python
# Create minutes columns with time in minutes
minutes = []

for i in df['Time']:
    minutes.append(i.hour *60 + i.minute + i.second /60)
    
    
df['Minutes'] = minutes
df['Minutes'] = df['Minutes'].round(decimals = 2)
```

```python
# Create Male dataframe
df_male = df[df['Gender'] == 'Male']
df_male.head()
```

```python
# create female dataframe
df_female = df[df['Gender'] == 'Female']
df_female.head()
```

```python
# Create males first, second, and third place dataframe
df_male_first = df_male[df_male['Place'] == 1]
df_male_second = df_male[df_male['Place'] == 2]
df_male_third = df_male[df_male['Place'] == 3]
```

```python
# Create females first, second, and third place dataframe
df_female_first = df_female[df_female['Place'] == 1]
df_female_second = df_female[df_female['Place'] == 2]
df_female_third = df_female[df_female['Place'] == 3]
```

```python
# Sort Dataframes in ascending order
df_male_first.sort_values(['Minutes'], axis =0, inplace = True)
df_male_second.sort_values(['Minutes'], axis =0, inplace = True)
df_male_third.sort_values(['Minutes'], axis =0, inplace = True)
df_female_first.sort_values(['Minutes'], axis =0, inplace = True)
df_female_second.sort_values(['Minutes'], axis =0, inplace = True)
df_female_third.sort_values(['Minutes'], axis =0, inplace = True)
```

```python
# sort dataframes Index
df_male_second.reset_index(inplace = True)
df_male_third.reset_index(inplace = True)
df_female_second.reset_index(inplace = True)
df_female_third.reset_index(inplace = True)
```

# What's the Male finishing time trend year to year ?

```python
# Average finishing time over the years for each position
plt.figure(figsize = [20,12])
plt.xticks(rotation = 90)
plt.yticks(np.arange(0,900,25))
sb.set_style('ticks')
sb.pointplot(data = df_male, y = 'Minutes', x = 'Year', hue = 'Place', ci = None)
plt.ylabel ('Finish Time in Min')
plt.xlabel ("Race Year")
#plt.xticks(np.arange(1978, 2019, 1))
plt.title('Year To Year Finsihing Times');
```

#### The finishing time for the Full Ironman Races is decreasing overtime
we can divid the time trends into three spanes:

* Year 1978 to 1989: Big decrease in the finishing time each year, the time gap between first to third place is very high. 

* Year 1989 to 2005: No improvements in the finishing time year to year except year 1996, also the gap between first to thirds place became more close than before.  

* Year 2005 to 2019: Small improvements in finsihing times over these years, however the competition is becoming more stronger and challenging.

Finsih time decrease year to year can be refered to many reasons including: 
1. Increased sport monitoring devices technology (Cycling Power meters, Heart rate monitoring, Cadence sensors)
2. Proffesional Training techniques based on collected metrics 

#### The Graphs shows that we can predict new finishing times in the future, however we don't expect dramatic improvements only slight improvements.
